```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from transformers import TFBertModel, TFBertForSequenceClassification, BertTokenizer
from tensorflow.keras.optimizers import Adam, SGD
from tensorflow.keras.losses import binary_crossentropy, categorical_crossentropy
```

```python
data = pd.read_csv('train_en.txt', sep='\t')
```

```python
data
```

```python
sentences = data['Sentence'].values.tolist()
labels = data['Style'].values.tolist()
```

```python
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
```

```python
tokenizer.encode_plus('This is a sentence')
```

```python
train_input_ids = []
train_attention_masks = []
```

```python
for sentence in sentences:
  result = tokenizer.encode_plus(sentence, max_length=15,
                                 truncation=True, pad_to_max_length=True)
  train_input_ids.append(result['input_ids'])
  train_attention_masks.append(result['attention_mask'])
```

```python
len(train_input_ids)
```

```python
len(train_input_ids[0])
```

```python
model = TFBertForSequenceClassification.from_pretrained('bert-base-uncased',
                                                        num_labels=2)
```

```python
model.compile(optimizer=Adam(learning_rate=0.01), loss=binary_crossentropy)
```

```python
model.fit([train_input_ids, train_attention_masks],
          [labels],
          epochs=5)
```
